```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Refund": ["Yes","No","No","Yes","No","No","Yes","No","No","No"],
    "MaritalStatus": ["Single","Married","Single","Married","Divorced","Married","Divorced","Single","Married","Single"],
    "TaxableIncome": [125,100,70,120,95,60,220,85,75,90],
    "Evade": ["No","No","No","No","Yes","No","No","Yes","No","Yes"]
}

df = pd.DataFrame(data)
df
```

```python
X = {
    "Refund": "No",
    "MaritalStatus": "Single",
    "TaxableIncome": 98
}
prior = df["Evade"].value_counts(normalize=True)
prior

```

```python
def conditional_prob(feature, value, target):
    subset = df[df["Evade"] == target]
    return len(subset[subset[feature] == value]) / len(subset)


p_refund_yes = conditional_prob("Refund", "No", "Yes")
p_refund_no = conditional_prob("Refund", "No", "No")

print(p_refund_yes, p_refund_no)

p_marital_yes = conditional_prob("MaritalStatus", "Single", "Yes")
p_marital_no = conditional_prob("MaritalStatus", "Single", "No")

print(p_marital_yes, p_marital_no)

```

```python
def gaussian_prob(x, mean, std):
    return (1 / (np.sqrt(2 * np.pi) * std)) * np.exp(-((x - mean) ** 2) / (2 * std ** 2))

    yes_income = df.loc[df["Evade"] == "Yes", "TaxableIncome"]
    no_income  = df.loc[df["Evade"] == "No",  "TaxableIncome"]
mean_yes, std_yes = yes_income.mean(), yes_income.std()
mean_no,  std_no  = no_income.mean(),  no_income.std()

mean_yes, std_yes, mean_no, std_no

p_income_yes = gaussian_prob(X["TaxableIncome"], mean_yes, std_yes)
p_income_no  = gaussian_prob(X["TaxableIncome"], mean_no, std_no)

p_income_yes, p_income_no

posterior_yes = prior["Yes"] * p_refund_yes * p_marital_yes * p_income_yes
posterior_no = prior["No"] * p_refund_no * p_marital_no * p_income_no

posterior_yes, posterior_no

if posterior_yes > posterior_no:
    print("Object X is classified as: Evade = YES")
else:
    print("Object X is classified as: Evade = NO")
```
