```python
import numpy as np
```

```python
points = np.array([
    [2,10],
    [2,5],
    [8,4],
    [5,8],
    [7,5],
    [6,4]
])

C1 = np.array([2,10])
C2 = np.array([5,8])

print("Initial C1:", C1)
print("Initial C2:", C2)

```

```python
def euclidean_distance(p, c):
    return np.sqrt(np.sum((p - c) ** 2))
for iteration in range(1, 3):
    cluster1 = []
    cluster2 = []

    # Assign points to nearest centroid
    for p in points:
        d1 = euclidean_distance(p, C1)
        d2 = euclidean_distance(p, C2)

        if d1 <= d2:   # <= to match homework expectation
            cluster1.append(p)
        else:
            cluster2.append(p)

    cluster1 = np.array(cluster1)
    cluster2 = np.array(cluster2)

    print(f"\nIteration {iteration}")
    print("Cluster 1:", cluster1)
    print("Cluster 2:", cluster2)

    # Update centroids
    C1 = np.mean(cluster1, axis=0)
    C2 = np.mean(cluster2, axis=0)

    print("Updated C1:", C1)
    print("Updated C2:", C2)
```
